```python
#3 area of triangle when sides are given
import math
a=float(input('enter the length of side a:'))
b=float(input('enter the length of side b:'))
c=float(input('enter the length of side c:'))
s=(a+b+c)/2
A=math.sqrt(s*(s-a)*(s-b)*(s-c))

print('the area of the triangle is:',A)
```

```python
# 6 area and circumference of the circle of given radius
import math
r=float(input('enter the radius of the circle:'))
c=2*math.pi*r
area=math.pi*r*r
print('the circumference of the circle is:',c)
print('the area of the circle is:',area)
```

```python
#9 finding average of ten numbers using while loop
sum=0
count=0
while(count<11):
    a = int(input('enter the value'))
    sum=sum+a
    count=count+1
    average= sum/count
print('average of ten numbers is:',average)
```

```python
#10 reversing a given number
a=int(input('enter the value of a:'))
rev=0
while(a!=0):
    digit=a%10
    rev=(rev*10)+digit
    a=a//10
print(rev)
```
